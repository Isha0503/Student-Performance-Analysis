# ⚙️ Feature Engineering

```python
df["Average"] = (df["G1"] + df["G2"] + df["G3"]) / 3
```

```python
df[["G1","G2","G3","Average"]].head()
```

```python
import numpy as np

df["Result"] = np.where(df["G3"] >= 10, "Pass", "Fail")
```

```python
df[["G3", "Result"]].head()
```

```python
def grade(g3):
    if g3 >= 18:
        return "A+"
    elif g3 >= 16:
        return "A"
    elif g3 >= 14:
        return "B"
    elif g3 >= 12:
        return "C"
    elif g3 >= 10:
        return "D"
    else:
        return "F"
```

```python
df["Grade"] = df["G3"].apply(grade)
```

```python
df[["G1", "G2", "G3", "Average", "Result", "Grade"]].head(10)
```

# 📊 Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of analyzing the dataset to discover patterns, relationships, and trends using statistical methods and visualizations.

## 📈 Graph 1: Average Final Grade by Gender

This graph compares the average final grade (G3) of male and female students.

```python
gender_avg = df.groupby("sex")["G3"].mean()

gender_avg
```

```python
# Interview Explanation
# Why did you use a Bar Chart?
# A bar chart is suitable because gender
# is a categorical variable, and we are comparing
# the average final grades between two categories.
plt.figure(figsize=(6,4))

plt.bar(
    gender_avg.index,
    gender_avg.values,
    color=["#ff69b4", "#4dabf7"]
)
plt.savefig(
    "/content/├── 04_gender_performance.png",
    dpi=300,
    bbox_inches="tight"
)
plt.title("Average Final Grade by Gender",
    fontsize=15,
    fontweight="bold")

plt.xlabel("Gender", fontsize=12)

plt.ylabel("Average Grade", fontsize=12)

for i, value in enumerate(gender_avg.values):
    plt.text(i, value+0.2, round(value,2), ha="center")
    plt.grid(alpha=0.3)

plt.show()
```

### 📌 Observation

- This graph compares the average final grade (G3) of male and female students.
- It helps identify whether one gender performs better academically.

## 🥧 Graph 2: Pass vs Fail Distribution

This graph shows the percentage of students who passed and failed the final examination.

```python
result = df["Result"].value_counts()

result
```

```python
#Interview Explanation

# Why did you use a Pie Chart?

# A pie chart is useful because it
# clearly represents the proportion of
# students who passed and failed.
plt.figure(figsize=(6,6))

plt.pie(
    result,
    labels=result.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["#4CAF50","#F44336"]
)
plt.savefig(
    "/content/05_pass_fail.png",
    dpi=300,
    bbox_inches="tight"
)
plt.title("Pass vs Fail", fontsize=15,
    fontweight="bold")
plt.grid(alpha=0.3)
plt.show()
```

### 📌 Observation

- The majority of students passed the final examination.
- Only a small percentage of students failed.

## 📊 Graph 3: Grade Distribution

This graph shows the number of students in each grade category.

```python
grade_count = df["Grade"].value_counts()

grade_count
```

```python
### Interview Explanation

#Why did you use a Bar Chart?

#A bar chart is appropriate because
#grades are categorical values (A+, A, B, C, D, F),
#and we want to compare the number of students in each category.
plt.figure(figsize=(7,5))

plt.bar(
    grade_count.index,
    grade_count.values,
    color="#FFA500"
)
plt.savefig(
    "/content/06_grade_distribution.png",
    dpi=300,
    bbox_inches="tight"
)
plt.title("Grade Distribution", fontsize=15,
    fontweight="bold")

plt.xlabel("Grade", fontsize=12)

plt.ylabel("Number of Students", fontsize=12)

for i, value in enumerate(grade_count.values):
    plt.text(i, value+2, value, ha="center")
plt.grid(alpha=0.3)
plt.show()
```

### 📌 Observation

- This graph shows how students are distributed across different grades.
- It helps identify the most common grade achieved by students.

## 📈 Graph 4: Study Time vs Final Grade

This graph analyzes whether students who spend more time studying achieve better final grades.

```python
### 🎤 Interview Explanation

#Why did you use a Line Chart?

#A line chart is used because study time is
#an ordered variable (1–4), making it easy to observe
#trends in average grades.
study_avg = df.groupby("studytime")["G3"].mean()

study_avg
```

```python
plt.figure(figsize=(7,5))

plt.plot(
    study_avg.index,
    study_avg.values,
    marker="o",
    linewidth=2,
    color="#6A5ACD"
)
plt.savefig(
    "/content/07_studytime_vs_grade.png",
    dpi=300,
    bbox_inches="tight"
)
plt.title("Study Time vs Average Final Grade", fontsize=15,
    fontweight="bold")
plt.xlabel("Study Time", fontsize=12)
plt.ylabel("Average Final Grade", fontsize=12)

plt.grid(True)
plt.grid(alpha=0.3)
plt.show()
```

### 📌 Observation

- Students who spend more time studying generally achieve better final grades.
- This suggests a positive relationship between study time and academic performance.

## 📉 Graph 5: Absences vs Final Grade

This graph examines how student absences affect final grades.

```python
### 🎤 Interview Explanation

#Why did you use a Scatter Plot?

#A scatter plot helps visualize the relationship between
#two numerical variables: absences and final grades.
plt.figure(figsize=(7,5))

plt.scatter(
    df["absences"],
    df["G3"],
    color="#20B2AA"
)
plt.savefig(
    "/content/08_absences_vs_grade.png",
    dpi=300,
    bbox_inches="tight"
)
plt.title("Absences vs Final Grade", fontsize=15,
    fontweight="bold")
plt.xlabel("Number of Absences", fontsize=12)
plt.ylabel("Final Grade", fontsize=12)

plt.grid(True)
plt.grid(alpha=0.3)
plt.show()
```

### 📌 Observation

- Students with fewer absences generally achieve higher grades.
- Higher absences may negatively affect academic performance.

# 📊 Overall Statistics

This section summarizes the overall performance of students.

```python
print("="*40)
print("STUDENT PERFORMANCE SUMMARY")
print("="*40)

print("Total Students :", len(df))
print("Average Final Grade :", round(df["G3"].mean(),2))
print("Highest Grade :", df["G3"].max())
print("Lowest Grade :", df["G3"].min())

pass_percentage = (df["Result"]=="Pass").mean()*100
fail_percentage = (df["Result"]=="Fail").mean()*100

print("Pass Percentage :", round(pass_percentage,2),"%")
print("Fail Percentage :", round(fail_percentage,2),"%")
```

### 📌 Observation

- This summary provides a quick overview of student performance.
- It highlights the average grade, highest grade, lowest grade, and pass/fail percentages.

# 🔥 Correlation Heatmap

The heatmap shows the relationship between numerical variables in the dataset.

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.savefig(
    "/content/12_heatmap.png",
    dpi=300,
    bbox_inches="tight"
)
plt.title("Correlation Heatmap", fontsize=15,
    fontweight="bold")

plt.show()
```

### 📌 Observation

- G1, G2, and G3 have a strong positive correlation.
- Students who perform well in earlier exams tend to perform well in the final exam.
